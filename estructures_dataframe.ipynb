**Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)**

Creem la columna Vol_Endarrerit. Si el temps de vol real és superior al temps previst, el resultat és True (que transformem a = 1 amb la funció map()). En cas contrari, és False (=0)

```python
delays1["Vol_Endarrerit"] = delays1["ActualElapsedTime"] > delays1["CRSElapsedTime"]
```

```python
delays1['Vol_Endarrerit'] = delays1['Vol_Endarrerit'].map(
                   {True:1 ,False:0})
```

```python
delays1[["Vol_Endarrerit","ActualElapsedTime","CRSElapsedTime"]]
```

Creem la columna Velocitat_Mitjana: Distance / (ActualElapsedTime/60). Transformem els minuts en hora per tenir km/h.

```python
delays1["Velocitat_Mitjana"] = round(delays1["Distance"] / (delays1["ActualElapsedTime"]/60))
```

```python
delays1[["Distance","ActualElapsedTime","Velocitat_Mitjana"]]
```

**Taula de les aerolínies amb més endarreriments acumulats**

```python
# Per extreure les 10 aerolínies amb més endarreriments acumulats, busquem el recompte de registres, agrupat per aerolinea. Seleccionem els 10 resultats TOP ordenats de més minuts acumulats a menys


aerolinies_endarr = delays1.groupby('UniqueCarrier')['Vol_Endarrerit'].count().sort_values(ascending=False) 
aerolinies_endarr.nlargest(10)
```

```python
# També ho podem representar en un gràfic de barres

aerolinies_endarr.sort_values(ascending=False).plot(kind='bar')
plt.title("Número d'endarreriments per aerolínia")
```
